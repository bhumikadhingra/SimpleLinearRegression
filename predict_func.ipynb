```python
import csv
import numpy as np


def cal_coef(x_list, y_list):
    x_mean = np.mean(x_list)
    y_mean = np.mean(y_list)
    
    xlxm = []
    xlxm2 = []
    ylym = []
    xmuly = []
    
    for i in range(len(x_list)):
        xlxm.append(x_list[i]-x_mean)
        ylym.append(y_list[i]-y_mean)
        xlxm2.append((x_list[i]-x_mean)*(x_list[i]-x_mean))
        xmuly.append(xlxm[i]*ylym[i])
#     xlxm2 = map(lambda x:x*x, xlxm)
#     print("successfully calculated the values")
    
    num = np.sum(xmuly)
    den = np.sum(xlxm2)
    
    beta_1 = num/den
    
    beta_0 = y_mean - beta_1*x_mean
    
#     print('\nbeta 1 value is ', beta_1)
#     print('\n\nbeta 0 value is ', beta_0)
    return [beta_0,beta_1]
    
        

def predict(x_predict, beta_values):                    # slr implementation 
    y_predict = beta_values[0] + beta_values[1]*x_predict
    
    return y_predict
    
def rss(actual_y, predicted_y):                         # calculating RSS value to calculate RSE
    rss_value = 0
    for i in range(len(actual_y)):
        error2 = (actual_y[i]-predicted_y[i])*(actual_y[i]-predicted_y[i])
        rss_value += error2
#     print("\nRss value is coming out to be ", rss_value)
    return rss_value

def rse(actual_y, predicted_y):                             # Calculating RSE value
    rss_val = rss(actual_y,predicted_y)
    n = len(actual_y)
    rse_val = (rss_val/n-2)**0.5
    
    
    print("\nRse value is ",rse_val)

def rsquare(actual_y, predicted_y):                                 # Calculating R-Square value
    tss = 0
    m = np.mean(actual_y)
    for k in range(len(actual_y)):
        var = (actual_y[k]-m)*(actual_y[k]-m)
        tss += var
    rss_valu = rss(actual_y, predicted_y)    
    r2 = (tss-rss_valu)/tss
    
    print('\nR Square value is ', r2)



file = open('geyser.csv', mode = 'r')
d = csv.reader(file)
x = []
y = []
for i in d:
    if i[1]!='eruptions':
        x.append(float(i[1]))
        y.append(float(i[2]))
# print("Values of x list")
# print(x)
# print('\n=========================================================\n')
# print("Values of y list")
# print(y)
betas = cal_coef(x,y)
ypredicted = []
for xp in x:
    ypredicted.append(predict(xp,betas))
    
# print("\nActual y values are ",y)
# print("\nPredicted y values are ",ypredicted)

rse(y,ypredicted)
rsquare(y,ypredicted)
# predict(x_pred, betas)
file.close() 
```
